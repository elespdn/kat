Proviamo a collazionare le lettere T3 e T196

```python
from collatex import *
collation = Collation()
testimone_P4 = open( "lettere/T196-part1-P4.txt", encoding='utf-8' ).read()
testimone_S2 = open( "lettere/T196-part1-S2.txt", encoding='utf-8' ).read()
testimone_S4 = open( "lettere/T196-part1-S4.txt", encoding='utf-8' ).read()
collation.add_plain_witness( "P4", testimone_P4 )
collation.add_plain_witness( "S2", testimone_S2 )
collation.add_plain_witness( "S4", testimone_S4 )
table = collate(collation, output='html2', layout='vertical')
print(table)
```
